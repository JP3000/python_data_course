```python
history_lt = history(symbol='SHSE.600050', frequency='1d', start_time='2021-01-01',  end_time='2022-02-24', fields='open, close, low, high, eob', adjust=ADJUST_PREV, df= True)
```

```python
x=[]
y=[]
for i in history_lt.iterrows():
    y.append(i[1].tolist()[0:4])
    x.append(i[1].values[4])
```

```python
k1=Kline()
k1.add_xaxis(x)
k1.add_yaxis("kline", y,itemstyle_opts=opts.ItemStyleOpts(color="pink", color0="black"),)
k1.set_global_opts(
        yaxis_opts=opts.AxisOpts(is_scale=True,splitarea_opts=opts.SplitAreaOpts(areastyle_opts=opts.AreaStyleOpts(opacity=1)
            )),
        xaxis_opts=opts.AxisOpts(is_scale=True),
        title_opts=opts.TitleOpts(title="Kline-基本示例"),datazoom_opts=[opts.DataZoomOpts(type_="inside"),opts.DataZoomOpts(type_="slider")]
    )
k1.render_notebook()
```

```python
from pyecharts.charts import Line
from pyecharts.faker import Faker

L=Line()
L.add_xaxis(x)
L.add_yaxis("中国联通", history_lt.high.tolist())
L.set_global_opts(title_opts=opts.TitleOpts(title="最高价"))
L.set_global_opts(
        yaxis_opts=opts.AxisOpts(is_scale=True,splitarea_opts=opts.SplitAreaOpts(areastyle_opts=opts.AreaStyleOpts(opacity=1)
            )),
        xaxis_opts=opts.AxisOpts(is_scale=True),
        title_opts=opts.TitleOpts(title="Kline-基本示例"),datazoom_opts=[opts.DataZoomOpts(type_="inside"),opts.DataZoomOpts(type_="slider")]
    )
L.render_notebook()
```
