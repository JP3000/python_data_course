```python
shape=[False,False,False]
for i in range(3,len(sz)):
    z0=sz.iloc[i-2,:]
    z1=sz.iloc[i-1,:]
    z2=sz.iloc[i,:]
    #print(z0,z1,z2)
    if all([(z0.close-z0.open)<-11,abs(z1.close-z1.open)<2,(z2.close-z2.open)>6,abs(z2.close-z2.open)>abs((z0.close-z0.open)*0.5),
           z1.close<z0.close,z1.open<z0.close,z1.close<z2.open,z1.open<z2.open,
           z0.close>z1.close]):
        shape.append(True)
    else:
        shape.append(False)
```
